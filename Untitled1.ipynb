```python
df.columns = (
    df.columns
    .str.replace('_', '', regex=False)
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
```

```python
print(df.columns)
```

```python
missions_per_year = (
    df.groupby('launch_year')
      .size()
      .reset_index(name='missions')
)
```

```python
missions_per_year.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(
    missions_per_year['launch_year'],
    missions_per_year['missions']
)
plt.title('ISRO Missions Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.show()
```

```python
success_rate = (
    df.assign(success=df['mission_status'] == 'Success')
      .groupby('launch_year')['success']
      .mean()
)

success_rate.plot(title='Mission Success Rate Over Time')
plt.show()
```

```python
df["launch_year"] = pd.to_datetime(df["launch_year"], dayfirst=True)
```

```python
df.set_index("launch_year", inplace=True)
```

```python
df.columns
```

```python
df.dtypes
```

```python
df["budget"] = (
    df["budget"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("crore", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

df["budget"] = pd.to_numeric(df["budget"], errors="coerce")
```

```python
df["budget"].dtype
```

```python
budget_trend = df.resample("Y")["budget"].sum()
budget_trend
```

```python
status_trend = df.groupby("mission_status").count()["name"]
status_trend
```

```python
status_trend.plot(kind="bar")
plt.title("Mission Status Distribution")
plt.xlabel("mission_status")
plt.ylabel("Number of Missions")
plt.show()
```

```python
rolling_budget = budget_trend.rolling(window=2).mean()
rolling_budget
```

```python
rolling_budget.plot()
plt.title("2-Year Rolling Average of ISRO Budget")
plt.xlabel("Year")
plt.ylabel("Budget (₹ Crore)")
plt.show()
```

```python
budget_trend.plot(marker='o')
plt.title("Original ISRO Budget Trend")
plt.xlabel("Year")
plt.ylabel("Budget (₹ Crore)")
plt.show()
```

```python
model = ARIMA(budget_trend, order=(1, 1, 1))
model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=5)
forecast
```

```python
forecast_years = pd.date_range(
    start=budget_trend.index[-1] + pd.DateOffset(years=1),
    periods=5,
    freq='Y'
)

forecast_series = pd.Series(forecast, index=forecast_years)
forecast_series
```

```python
plt.figure(figsize=(8,5))

plt.plot(budget_trend, label="Actual Budget", marker='o')
plt.plot(forecast_series, label="Forecast Budget", marker='o', linestyle='--')

plt.title("ISRO Budget Forecast using ARIMA")
plt.xlabel("Year")
plt.ylabel("Budget (₹ Crore)")
plt.legend()
plt.show()
```
